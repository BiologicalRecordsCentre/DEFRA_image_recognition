```python
from os import listdir
from os.path import join, isfile, isdir
import pandas as pd

from ImageClassification import ImageClassification

from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.optimizers import Adam

# Warnings about data rate limits
# IOPub data rate exceeded.
# https://github.com/jupyter/notebook/issues/2287
# --NotebookApp.iopub_data_rate_limit (increase limit)
# jupyter notebook --generate-config
```

```python
species_home_directory = '/media/david/DataOne/data-sets/butterfly-species/butterflies/'
#species_home_directory = '/home/dciar86/GitHub/deep-learning/image-classification/butterflies/'
verified_flag = 'Verified'
validation_flag = 'Unverified'
# List all the directories, these become the classes
species_directories = [dir for dir in listdir(species_home_directory) if isdir(join(species_home_directory, dir))]

# Create the base dataframe
butterfly_df = pd.DataFrame()
validation_df = pd.DataFrame()

for current_species in species_directories:
    # Get the verified images
    images = [f for f in listdir(join(species_home_directory, current_species, verified_flag)) 
             if isfile(join(species_home_directory, current_species, verified_flag, f))]
    # Only for cpu testing
    images = images[0:5000]
    species_class = [current_species] * len(images)
    species_directory = [join(species_home_directory, current_species, verified_flag)] * len(images)
    dict_rep = {'image_id': images, 'directory': species_directory, 'class': species_class}
    df_rep = pd.DataFrame(dict_rep)
    butterfly_df = butterfly_df.append(df_rep, ignore_index=True) 

for current_species in species_directories:
    # Get the verified images
    images = [f for f in listdir(join(species_home_directory, current_species, validation_flag)) 
             if isfile(join(species_home_directory, current_species, validation_flag, f))]
    # Only for cpu testing
    images = images[0:5000]    
    species_class = [current_species] * len(images)
    species_directory = [join(species_home_directory, current_species, validation_flag)] * len(images)
    dict_rep = {'image_id': images, 'directory': species_directory, 'class': species_class}
    df_rep = pd.DataFrame(dict_rep)
    validation_df = validation_df.append(df_rep, ignore_index=True) 

print("Length of train/test data: {}, length of validation data: {}".format(butterfly_df.shape[0], validation_df.shape[0]))
```

```python
butterfly_classifier = ImageClassification('id1',
                                           butterfly_df, 
                                           validation_df,
                                           '/media/david/DataOne/data-sets/butterfly-species/output',  
                                           32,
                                           0.25,
                                           1,
                                           True,
                                           True)
```

```python
inception_resnet_top_model_basic = Sequential()
inception_resnet_top_model_basic.add(Dense(butterfly_classifier.class_counts.shape[0], activation='softmax', input_shape=(butterfly_classifier.model_dense_pooled_dimensions.get('inception-resnet'))))
inception_resnet_top_model_basic.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model_dict = {'BasicV3Res': [inception_resnet_top_model_basic, 'inception-resnet', 500]}

butterfly_classifier.train_provided_top_models(model_dict)
```
